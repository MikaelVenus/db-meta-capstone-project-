Query with table JOIN

For the third and final task, Little Lemon need you to return specific details from your database. They require the full name and contact details for every customer that has placed an order greater than $60 for a promotional campaign. 

Sample Data Script

```python
# -- Step 1: Cleanup existing data
# #DELETE FROM Orders;
# #DELETE FROM Menus;
# #DELETE FROM MenuItems;
# #DELETE FROM Customers;

# -- Step 2: Insert Menu Items (The components)
 
# -- Step 3: Insert Menus (The collections)
# #INSERT INTO Menu(MenuID, MenuItemsID, MenuName, Cusine)
# #VALUES
# #(10, 1, 'Mediterranean Delights', 'Greek'),
# #(11, 2, 'Italian Classic', 'Italian'),
# #(12, 3, 'Middle Eastern Feast', 'Turkish');

# -- Step 4: Insert Customers

# -- Step 5: Insert Orders
# INSERT INTO Orders (OrderId, Quantity, TotalCost, CustomerID, MenuId)
# VALUES
# (1, 2, 85.00, 1, 10), -- Vanessa ordered Greek ($85)
# (2, 3, 75.00, 2, 11); -- Marcos ordered Italian ($75)
```

```python
# Step 1, 2, & 3: Define the Join Query as a string
join_query = """
SELECT 
    Customers.FullName, 
    Customers.ContactNumber, 
    Customers.Email, 
    Orders.TotalCost
FROM Customers
INNER JOIN Orders 
ON Customers.CustomerID = Orders.CustomerID
WHERE Orders.TotalCost > 60;
"""

print("Running promotional campaign query...")

# Execute the query
cursor.execute(join_query)

# Fetch the results
results = cursor.fetchall()

# Display the findings
print("-" * 30)
print("{:<20} | {:<15} | {:<10}".format("Full Name", "Contact", "Total Cost"))
print("-" * 30)

for row in results:
    fullname, contact, email, cost = row
    print("{:<20} | {:<15} | ${:<10}".format(fullname, contact, cost))

print("-" * 30)
```
